```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('G:/My Drive/4. Projetos/Data Analytics/1. Spotify/archive/tracks.csv')
df_tracks.head()
```

```python
# Check if there is any empty cells - for this i'm going to use the method "isnull" on the Pandas library
pd.isnull(df_tracks).sum()
```

```python
# I'm going to delete the rows we can't know the name of the song
df_tracks = df_tracks.dropna()
```

```python
# In order to know the number of rows, columns and datatypes in the dataset - I'm going to use the "info" method
df_tracks.info()
```

```python
# Get the 10 least popular songs on Spotify - I sorted the popularity column by ascending in order to get it
sorted_df = df_tracks.sort_values('popularity', ascending=True).head(10)
sorted_df
```

```python
df_tracks.describe().transpose()
```

```python
# Now the top 10 songs with popularity above 90
most_popular = df_tracks.query('popularity>90', inplace=False).sort_values('popularity', ascending=False).head(10)
most_popular[:10]
```

```python
# Visualization of the top 10 most popular Songs
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
sns.barplot(y='name', x='popularity', data = most_popular).set(title='Top 10 most popular songs')
```

```python
# Now the top 10 most danceable songs with popularity above 90
most_danceable = df_tracks.query('popularity>90', inplace=False).sort_values('danceability', ascending=False).head(10)
most_danceable
```

```python
# Now the top 10 most danceable songs with popularity above 90
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
sns.barplot(y='name', x='danceability', data = most_danceable).set(title='Top 10 Popular songs most Danceable ')
```

```python
# Setting release_date as index in the main df
df_tracks.set_index('release_date',inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
```

```python
df_tracks.head()
```

```python
df_tracks.head()
```

```python
# Change the duration from ms to sec, so I'm going to divide by 1000.
df_tracks['duration']= df_tracks['duration_ms'].apply(lambda x:round(x/1000))
df_tracks.drop('duration_ms', inplace=True,axis=1)
```

```python
df_tracks.duration.head()
```

```python
# Now I want to know which are the most popular songs on the first year of pandemic (2020 - covid)
df_tracks_2022 = df_tracks.loc['2020-01-01' : '2020-12-31'].sort_values('popularity', ascending=False)
df_tracks_2022.head(10)
```

```python
corr_df = df_tracks.drop(['key','mode','explicit'], axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g",vmin=-1, vmax=1, center=0, cmap="inferno",linewidths=1, linecolor='black')
heatmap.set_title("Correlation HeatMap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
# Here we can see a positive correlation between the two variables, meaning that when the loudness in a song increases the energy increases as well and vice versa
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y = 'loudness', x= 'energy', color = 'c'). set(title= 'Loudness vs Energy Correlation')
```

```python
# Here we can see a negative correlation between the two variables, meaning that when the popularity in a song increases the energy decreases and vice versa
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y = 'popularity', x= 'acousticness', color = 'b'). set(title= 'Popularity vs Acousticness Correlation')
```

```python
# Now I want to see a visualization of the number of songs released each year
# For that first I need to create "year" column
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title="Number of Songs per year")
```

```python
# Creating a visualization to analyse the duration of the songs over the years
total_dr = df_tracks.duration
sns.set_style(style='whitegrid')
fig_dims = (10,5)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x=years, y=total_dr, ax = ax).set(title='Year vs Duration')
plt.xticks(rotation=60)
```

Importing a new Spotify Features Dataset with more data

```python
df_genre = pd.read_csv('G:/My Drive/4. Projetos/Data Analytics/1. Spotify/archive/SpotifyFeatures.csv')
df_genre.head()
```

```python
plt.title('duration of the Songs by Genres')
sns.color_palette('rocket', as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data=df_genre)
plt.xlabel('Duration in ms')
plt.ylabel('Genres')
```

```python
# Now, to see the top 5 genres based on popularity
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
famous = df_genre.sort_values('popularity', ascending=False).head(10)
sns.barplot(y='genre', x='popularity', data = famous).set(title='Top 5 Genres by Popularity')
```

```python
# Since I'm a fan of Rap music I'm going to find the most danceable and most popular songs in this genre, for this I'm going to use the "speechiness" column.
# "Speechiness detects the presence of spoken words in a track, the more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value.
df_hr = df_genre[df_genre['genre'].isin(['Rap'])]
df_hr.head()
```

```python
# Most danceable and most popular songs in Rap
most_danceable_hr = df_hr.query('popularity>90', inplace=False).sort_values('danceability', ascending=False).head(10)
most_danceable_hr.head(10)
```

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
sns.barplot(y='track_name', x='danceability', data = most_danceable_hr).set(title='The most danceable and popular songs in Rap')
```
